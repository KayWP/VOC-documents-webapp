```python
import pandas as pd
```

```python
df = pd.read_csv('data/file_to_render.csv')
df = df.head(1000)
df['transcription_viewer_link'] = df['transcription_viewer_link'].apply(lambda x: f'<a href="{x}" target="_blank">{x}</a>')
```

```python
# Generate interactive HTML table with DataTables
html = f"""
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Data Table</title>
    
    <!-- DataTables CSS -->
    <link rel="stylesheet" href="https://cdn.datatables.net/1.11.5/css/jquery.dataTables.min.css">
    
    <!-- jQuery -->
    <script src="https://code.jquery.com/jquery-3.6.0.min.js"></script>
    
    <!-- DataTables JS -->
    <script src="https://cdn.datatables.net/1.11.5/js/jquery.dataTables.min.js"></script>
    
    <style>
        body {{
            font-family: Arial, sans-serif;
            margin: 20px;
            padding: 20px;
        }}
        
        #myTable {{
            width: 100% !important;
        }}
    </style>
</head>
<body>
    <h1>Data Table</h1>
    
    {df.to_html(table_id='myTable', index=False, escape=False)}
    
    <script>
        $(document).ready(function() {{
            $('#myTable').DataTable({{
                pageLength: 25,
                order: [[0, 'asc']],
                responsive: true
            }});
        }});
    </script>
</body>
</html>
"""

# Write HTML to file
with open('output.html', 'w', encoding='utf-8') as f:
    f.write(html)

print("HTML table generated successfully: output.html")
```
